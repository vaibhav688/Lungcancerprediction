```python
import pandas as pd
data=pd.read_csv('/kaggle/input/lung-cancer/survey lung cancer.csv')
```

```python
data
```

```python
data['LUNG_CANCER']=data['LUNG_CANCER'].map({'NO':0, "YES":1})
```

```python
data
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
input_cols=list(data.columns)[0:-1]
target_cols='LUNG_CANCER'
```

```python
input_cols
```

```python
inputs_df=data[input_cols]
target=data[target_cols]
```

```python
inputs_df
```

```python
target
```

```python
inputs_df.info(10)
```

```python
inputs_df['GENDER']=inputs_df['GENDER'].map({'M':1,'F':2})
```

```python
inputs_df
```

```python
inputs_df.info(10)
```

```python
from sklearn.model_selection import train_test_split
train_input,val_input,train_target,val_target=train_test_split(inputs_df,target,test_size=0.2,random_state=42)
```

```python
train_input
```

```python
import numpy as np
numeric_col=train_input.select_dtypes(include=np.number).columns.tolist()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit(inputs_df)
train_input[numeric_col]=scaler.transform(train_input[numeric_col])
val_input[numeric_col]=scaler.transform(val_input[numeric_col])
```

```python
train_input
```

```python
X_train=train_input[numeric_col]
X_val=val_input[numeric_col]
```

```python
from sklearn.linear_model import LogisticRegression
mod=LogisticRegression().fit(train_input[numeric_col],train_target)
```

```python
mod.score(X_train,train_target),mod.score(X_val,val_target)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier().fit(train_input[numeric_col],train_target)
```

```python
tree.score(X_train,train_target),tree.score(X_val,val_target)
```
